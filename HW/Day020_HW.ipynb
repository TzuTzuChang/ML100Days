```python
#random_sample可生成[0, 1)的隨機數, 1000行, 3列
#再把隨機數-0.5*2, 可使數值調整範圍至[-1, 1)
matrix = (np.random.random_sample(size = (1000, 3)) - 0.5)*2
matrix
```

```python
nrow = 1000
ncol = 3

#隨機給予0, 1, 2三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()
```

```python
#繪製seborn進階Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

#上半部為scatter
grid.map_upper(plt.scatter, alpha = 0.2)
#對角線畫histogram
grid.map_diag(plt.hist)
#下半部放density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
#np.random.randn-隨機產生常態分配, 1000行, 3列
matrix = np.random.randn(1000, 3)
matrix
```

```python
#圖形驗證資料是否為常態分佈
plt.hist(matrix)
```

```python
nrow = 1000
ncol = 3

#隨機給予0, 1, 2三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

#繪製seborn進階Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

#上半部為scatter
grid.map_upper(plt.scatter, alpha = 0.2)
#對角線畫histogram
grid.map_diag(plt.hist)
#下半部放density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
