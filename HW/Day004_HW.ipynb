#### 截取部分資料

```python
#DataFrame截取資料方法有三種
#法1.中括號 ([ ], square bracket)
#法2.iloc
#法3.loc
#中括號的使用方式很類似python在其他資料型態, 如list的使用方式
#中括號最大的缺點是無法同時選擇row和column
#iloc和loc則是Pandas獨有的截取資料方法
#iloc是透過資料的index來選取
#而loc則是利用label來選取
```

```python
#法1-中括號
#所有欄都要, 截取10-15列
print(app_train[10:15])
print(type(app_train[10:15]))
```

```python
#所有列都要, 但只要特定的1欄
print(app_train['SK_ID_CURR'])
print(type(app_train['SK_ID_CURR']))
```

```python
#所有列都要, 但只要特定的2欄
#若選多欄需要二個中括號[[]]
print(app_train[['SK_ID_CURR', 'FLAG_OWN_REALTY']])
print(type(app_train[['SK_ID_CURR', 'FLAG_OWN_REALTY']]))
```

```python
#法2-iloc, 用資料的index擷取資料
#df.iloc[]-用index擷取資料特定欄或特定列的資料
#所有欄位都要, 但只挑2-5列
app_train.iloc[[2,3,4,5]]
#也可寫成app_train.iloc[[2,3,4,5], :]
```

```python
#所有列都要, 但只截取0,2,3欄
app_train.iloc[:, [0,2,3]]
```

```python
#截取1-3列與0-3欄
app_train.iloc[[1,2,3], [0,2,3]]
```

```python
#法3-loc, 用資料的label擷取資料
#df.loc[]-用label截取資料特定欄或特定列的資料
#所有欄位都要, 但只挑'100006', '100008'列
app_train.loc[(app_train['SK_ID_CURR'] == 100006) | (app_train['SK_ID_CURR'] == 100008)]
```

```python
#所有欄位都要, 但只挑'100006'這個資料列
app_train.loc[(app_train['SK_ID_CURR'] == 100006)]
```

```python
#所有列都要, 但只截取'SK_ID_CURR', 'CODE_GENDER'欄
app_train.loc[:, ['SK_ID_CURR', 'CODE_GENDER']]
```

```python
#指定2:5列, 與'SK_ID_CURR', 'CODE_GENDER'欄
app_train.loc[2:5, ['SK_ID_CURR', 'CODE_GENDER']]
```

```python
#指定'100006','100008'列, 與'SK_ID_CURR', 'CODE_GENDER'欄
app_train.loc[(app_train['SK_ID_CURR'] == 100006) | (app_train['SK_ID_CURR'] == 100008), ['SK_ID_CURR', 'CODE_GENDER']]
```
